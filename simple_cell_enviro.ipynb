```python
def create_environment(w, h):
    #define empty occupancy grid
    envr_matrix = [[0 for x in range(w)] for y in range(h)] 
    return envr_matrix


def create_perimeter(centroid, w, h):
    #draw rectangle
    x = []
    y = []
    
    for i in range(w + 1):
        x.append((centroid[0] - (w/2) + i))
        y.append(centroid[1] + (h/2))
        x.append(centroid[0] - (w/2)+i)
        y.append(centroid[1] - (h/2))
        
    for i in range(h+1):
        x.append((centroid[0] - (w/2)))
        y.append((centroid[1] + (h/2))-i)
        x.append((centroid[0] + (w/2)))
        y.append((centroid[1] + (h/2))-i)
        
    envr = []
    envr.append((x, y) )
  
    return x, y, envr

def create_obstacle(w,h,centroid,shape):
    if shape == "rectangle":
        for i in range(w + 1):
            for j in range(h+1):
                x.append((centroid[0] - (w/2) + i))
                y.append(centroid[1] + (h/2)-j)
    return x, y

def add_obstacle(x_obst, y_obst,x, y):
    for i in range(len(x_obst)):
        x.append(x_obst[i])
        y.append(y_obst[i])


    return x, y

def create_robot(start_point, size):
    
    robot_x = []
    robot_y = []
    for i in range(size + 1):
        for j in range(size+1):
            robot_x.append((start_point[0] - (size/2) + i))
            robot_y.append(start_point[1] + (size/2)-j)
    return robot_x,robot_y

def fill_occupancy_grid(occupancy_grid, x, y):
    print("len(x) ",len(x))
    print("len(y) ",len(y))
    print(len(occupancy_grid[0]))
    
    for i in range(len(x)):
        occupancy_grid[int(y[i])][int(x[i])] =1
    
    return occupancy_grid

    
```
